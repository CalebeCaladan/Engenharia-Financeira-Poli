### Exercício Hidrelétrica em Moçambique

Uma empresa que acabou de ganhar uma concessão para operar uma hidrelétrica em Moçambique, com a perspectiva de obter um fluxo de caixa livre (FCn) em Dólar de $100 MM ao ano nos próximos 30 anos, encomendou-lhe o seguinte estudo:

Continuação:

(i) Se a alternativa para compensar o atraso for aumentar o valor do FC livre, de sorte que sejam maiores que $100MM anuais, mantendo o período original da concessão em 30 anos no total, quanto este FC deveria valer para compensar o atraso dos 2 anos, (c/ 3%)?

(j) Se a alternativa para compensar o atraso for aumentar a quantidade de anos de concessão para explorar o empreendimento, usufruindo do FC livre propostos, quanto tempo em anos seria necessário na situação de uma taxa de desconto de 3%aa?

(k) Vc consegue imaginar outras alternativas que poderiam ser propostas para compensar este atraso?

#### Resolução

```python
#  (b)
import numpy_financial as npf
import numpy as np
import pandas as pd
from tabulate import tabulate

anos = np.arange(0,31)
fc = np.append([-500.0], np.repeat(100.0, 30))

df = pd.DataFrame({'Anos': anos, 'FCn': fc})

rate = 3/100
vpl = npf.npv(rate, fc).round(5)

df['Taxa 3%'] = df['FCn'].values / (1+(3/100))**df['Anos'].values

df = df.round(2)

print(f'VPL do fluxo sem atraso: {vpl}')
```

### Exercício 1
Calcule o VPL e na sequência calcule a TIR de um projeto que tem 100 milhões de investimento inicial, (pois os investidores compraram por 80 mi e, no mesmo instante, gastaram mais 20 mi), se ele for vendido no período seguinte por 145 mi, considerando uma taxa de desconto de 10%.

#### Resolução

```python
# (Exercício 1)

import numpy as np
import pandas as pd
from tabulate import tabulate

anos = np.arange(0,2)
fc = np.array([-100.0, 145.0])

df = pd.DataFrame({'Anos': anos, 'FCn': fc})
df['Taxa 10%'] = df['FCn'].values / (1+(10/100))**df['Anos'].values

df = df.round(2)

rate = 10/100
vpl = npf.npv(rate, fc).round(5)

tir = round(npf.irr(df['FCn'].values), 4)

print(tabulate(df, headers='keys', tablefmt='psql', showindex=False))
print(f'VPL do Fluxo: {vpl}')
print(f'TIR do Fluxo: {tir:.2%}')
```

### Exercício 2
Qual o VPL e qual a TIR de um projeto de geração de energia em que os investidores pagaram a vista 180 mil, e depois de obter um fluxo de caixa livre de 30 mil no período seguinte com a venda da energia, venderam o proj. por 210 mil para outros interessados? (Considere a taxa de desconto de 10% ao período).

#### Resolução

```python
# (Exercício 2)

import numpy as np
import pandas as pd
from tabulate import tabulate

anos = np.arange(0,2)
fc = np.array([-180.0, 210.0+30.0])

df = pd.DataFrame({'Anos': anos, 'FCn': fc})
df['Taxa 10%'] = df['FCn'].values / (1+(10/100))**df['Anos'].values

df = df.round(2)

rate = 10/100
vpl = npf.npv(rate, fc).round(5)

tir = round(npf.irr(df['FCn'].values), 4)

print(tabulate(df, headers='keys', tablefmt='psql', showindex=False))
print(f'VPL do Fluxo: {vpl}')
print(f'TIR do Fluxo: {tir:.2%}')
```

1 - Seguindo os três passos da análise de eventos que alteram o equilíbrio entre oferta e demanda, ilustre graficamente o que ocorre com o mercado de bens duráveis (carros, televisores, celulares, etc), tendo em vista o aumento do desemprego provocado pela pandemia do Coronavírus, e responda o que deverá ocorrer com preço (P), quantidade (Q) e receita (R) no novo equilíbrio.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

sns.set(style="darkgrid")

demand1 = np.arange(1000, 200, -125) # Eixo X
price1 = np.arange(500, 1200, 100) # Eixo Y

#demand2 = np.arange(1600, 800, -125)
#price2 = np.arange(400, 1100, 100)

supply = np.arange(1300, 500, -50) # Eixo X
price_supply = np.arange(1200, 400, -50) # Eixo Y

supply2 = np.arange(1200, 400, -50) # Eixo X
price_supply2 = np.arange(1300, 500, -50) # Eixo Y

plt.figure(figsize=(10,6))

# Plotando as curvas.
plt.plot(demand1, price1, 'dodgerblue', supply, price_supply, 'grey', supply2, price_supply2, 'grey', linewidth=5.0)

# Delimitando.
plt.axis((0, 1800, 300, 1400))

# labels
plt.xlabel('Quantidade (Q)')
plt.ylabel('Preço (P)')

# Anotações
plt.annotate('Demanda', xy=(230, 1120))
plt.annotate('O1', xy=(1300, 1210))
plt.annotate('O2', xy=(1210, 1300))

plt.title('Mercado Bens Duraveis')

plt.show()
```

Resposta:
Tendo em vista a diminuição do poder aquisitivo da população, a quantidade de itens vendidos deve cair e a variável preço deve subir. A receita se manteria a mesma.
