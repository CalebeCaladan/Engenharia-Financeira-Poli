### Exercício 1 

Você tem um "amigo da onça". Ele cobra 1% ao dia pelo dinheiro que ele empresta aos outros. Se você pegar uma grana com ele e se esquecer de paga-lo de volta pelo período de um ano, quanto será a taxa total de juros neste período, referente ao que você terá que pagar pelo seu empréstimo?

- [ ] 365% ao Ano
- [ ] Entre 500 e 1000% no ano
- [X] Mais que 1000% no ano

#### Resolução

$$
FV = PV (1 + R) ^ n
$$
<br>
$$
FV = 1 (1 + (1/100))^{365}
$$
<br>
$$
FV = (37.78 - 1) * 100
$$
<br>
$$
FV = 3678.34\%
$$

```python
#  Resolução usando as bibliotecas.
import numpy_financial as npf

r = 1/100
n = 365
pv = -1
fv = npf.fv(r,n,0,pv)
fv = (fv - 1)
print(f'Taxa ao final do periodo: {fv:.2%}')
```

### Exercicio 2

Uma empresa que acabou de ganhar uma concessão para operar uma hidrelétrica em Moçambique, com a perspectiva de obter um fluxo de caixa livre (FCn) em Dólar de $100 MM ao ano nos próximos 30 anos, encomendou-lhe o seguinte estudo:

(a) Ela gostaria de saber qual é o Payback desse projeto, caso ela pague $ 500 MM de investimento inicial pelo direito de operar essa hidrelétrica e de obter o fluxo de caixa livre indicado acima.

(b) Ela gostaria de saber também qual é o Valor Presente Líquido (VPL) do projeto se os fluxos de caixa forem descontados às taxas de 3%, depois de 9% e depois de 15%, considerando o investimento inicial de $ 500 MM, conforme já indicado. (Use o Excel).

(c) Ela pede que vc represente, num único gráfico de barras, o Valor Presente (VP) dos FCn descontados a 3%, a 9%, e a 15%. (No eixo vertical estará o VP dos FCn em dólares,
e, no eixo horizontal, o tempo em anos. Para cada ano haverá três barras, conforme ilustrado em classe. Represente apenas da parte positiva desses três fluxos, ou seja, não inclua no seu gráfico os “- $500” referentes ao invest. inicial). (Use o Excel)

#### Resolução

```python
#  (a) Se o invesitmento inicial for de $ 500 MM o Payback será de 5 anos.

investimento = 500
fluxo_caixa = 100

payback = investimento / fluxo_caixa

print(f'Se o invesitmento inicial for de $ 500 MM o Payback será de {payback} anos.')
```

```python
#  (b)
import numpy as np
import pandas as pd
from tabulate import tabulate

anos = np.arange(0,31)
fc = np.append([-500.0], np.repeat(100.0, 30))

df = pd.DataFrame({'Anos': anos, 'FCn': fc})

df['Taxa 3%'] = df['FCn'].values / (1+(3/100))**df['Anos'].values
df['Taxa 9%'] = df['FCn'].values / (1+(9/100))**df['Anos'].values
df['Taxa 15%'] = df['FCn'].values / (1+(15/100))**df['Anos'].values

df = df.round(2)

print(tabulate(df, headers='keys', tablefmt='psql', showindex=False))
```

```python
#  (c)
import matplotlib.pyplot as plt
import seaborn as sns

#plt.style.use('seaborn')
sns.set(style="darkgrid")

y1 = df['Taxa 3%'][1:].to_list()
y2 = df['Taxa 9%'][1:].to_list()
y3 = df['Taxa 15%'][1:].to_list()

width = 0.30
x = np.arange(len(y1))

plt.figure(figsize=(16, 6))

plt.bar(x-0.30, y1, width)
plt.bar(x, y2, width)
plt.bar(x+0.30, y3, width)

plt.xlabel('Tempo em Anos')
plt.ylabel('Fluxo Caixa Descontado em MM $')
plt.title('Demostração Grafica');
```
